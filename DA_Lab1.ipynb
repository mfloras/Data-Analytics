## Data Analytics - Lab 1 - Random data visualisation and sampling

```python
from numpy import array
from numpy.random import uniform, normal, beta, multivariate_normal
import matplotlib.pyplot as plt
from numpy import linspace
import seaborn as sns
```

### Exercise 2

```python
# uniform distribution
uni = uniform(0, 1, 1000)

# normal distribution with mean 0 and standard deviation 1
normal1 = normal(0, 1, 1000)

# normal distribution with mean 0 and standard deviation 10
normal2 = normal(0, 10, 1000)

# beta distribution for the pairs of parameters 
beta_d = beta(array([1, 2, 3]), array([1, 2, 9]), (1000, 3))

#  two-dimensional normal distributions
mean = [0, 0]
cov1 = array([[1, 0], [0, 1]])
cov2 = array([[2, 1], [1, 2]])
e1 = multivariate_normal([0,0], cov1, 10000)
e2 = multivariate_normal(mean, cov2, 10000)
```

### Exercise 3

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

axs[0,0].hist(uni, density=True, bins=20)
axs[0,0].set_title("Unform distribution")

axs[0,1].hist(normal1, density=True, bins=20)
axs[0,1].set_title("Normal distribution $\mu=0$, $\sigma=1$")

axs[0,2].hist(normal2, density=True, bins=20)
axs[0,2].set_title("Normal distribution $\mu=0$, $\sigma=10$")

axs[1,0].hist(beta_d[:, 0], density=True, bins=20)
axs[1,0].set_title("Beta distribution $\\alpha=1$, $\\beta=1$")

axs[1,1].hist(beta_d[:, 1], density=True, bins=20)
axs[1,1].set_title("Beta distribution $\\alpha=2$, $\\beta=2$")

axs[1,2].hist(beta_d[:, 2], density=True, bins=20)
axs[1,2].set_title("Beta distribution $\\alpha=3$, $\\beta=9$")

plt.show()
```

The appearance of above plots is similar to analitcal plots from web sources. In order to improve their accuracy it is needed to increase $bins$ parameter.

### Exercise 4

```python
# using hist2d
fig, axs = plt.subplots(2, 4, figsize=(16, 8))
fig.suptitle("Two-dimensional normal distribution")

axs[0, 0].hist2d(e1[:, 0], e1[:, 1], bins=20)
axs[0, 0].set_title("1 covariance matrix, bins = 20")
axs[0, 1].hist2d(e2[:, 0], e2[:, 1], bins=20)
axs[0, 1].set_title("2 covariance matrix, bins = 20")

axs[0, 2].hist2d(e1[:, 0], e1[:, 1], bins=40)
axs[0, 2].set_title("1 covariance matrix, bins = 40")
axs[0, 3].hist2d(e2[:, 0], e2[:, 1], bins=40)
axs[0, 3].set_title("2 covariance matrix, bins = 40")

axs[1, 0].hist2d(e1[:, 0], e1[:, 1], bins=60)
axs[1, 0].set_title("1 covariance matrix, bins = 60")
axs[1, 1].hist2d(e2[:, 0], e2[:, 1], bins=60)
axs[1, 1].set_title("2 covariance matrix, bins = 60")

axs[1, 2].hist2d(e1[:, 0], e1[:, 1], bins=80)
axs[1, 2].set_title("1 covariance matrix, bins = 80")
axs[1, 3].hist2d(e2[:, 0], e2[:, 1], bins=80)
axs[1, 3].set_title("2 covariance matrix, bins = 80")

plt.show()
```

```python
# tryng histplot from seaborn
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
fig.suptitle("Two-dimensional normal distribution using seaborn")

sns.histplot(data=e1, x=e1[:, 0], y=e1[:, 1], ax=axs[0])
axs[0].set_title("1 covariance matrix")
sns.histplot(data=e1, x=e1[:, 0], y=e1[:, 1], ax=axs[1])
axs[1].set_title("2 covariance matrix")

plt.show()
```

For greater values of bins, plots are more accurate.

### Exercise 5

```python

alpha = normal(2, 0.1, 1000)
beta = normal(0.2, 0.2, 1000)
x = normal(0, 3, 1000)
sigma = abs(normal(0, 1, 1000))
u = alpha + beta@x
y = normal(u, sigma, 1000)
```

### Exercise 6

```python
x_lin = linspace(-6, 6, num=1000)
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x, y)
for i in range(1000):
    ax.plot(x_lin, alpha[i]+beta[i]*x_lin, c='r')

plt.show()
```

### Conclusions

Using numpy library it is possible to generate random data about a given distribution. To visualize prepared data, the best way is using libraries like matplotlib or seaborn.
